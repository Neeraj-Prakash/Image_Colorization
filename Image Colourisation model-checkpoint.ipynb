```python
import pickle
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, UpSampling2D, BatchNormalization, Activation, Flatten, Conv2D, Conv2DTranspose
import numpy as np

x = pickle.load(open("Grey.pickle","rb"))
y = pickle.load(open("LAB.pickle","rb"))

x= x/255.0
y =np.array(y).reshape(-1,256,256,3)
```

```python
model = Sequential(name= 'colorizer')

#conv1
model.add(Conv2D(64,(3,3), input_shape= x.shape[1:], strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(64, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(BatchNormalization())

#conv2
model.add(Conv2D(128, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(128, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(BatchNormalization())
          
#conv3
model.add(Conv2D(256, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(256, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(256, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(BatchNormalization())
          
#conv4
model.add(Conv2D(512, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(512, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(512, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(BatchNormalization())

#conv5
model.add(Conv2D(512, (3,3), dilation_rate=2, strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(512, (3,3), dilation_rate=2, strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(512, (3,3), dilation_rate=2,  strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(BatchNormalization())
    
#conv6
model.add(Conv2D(512, (3,3), dilation_rate=2, strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(512, (3,3), dilation_rate=2, strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(512, (3,3), dilation_rate=2, strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(BatchNormalization())

#conv7
model.add(Conv2D(512, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(512, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(512, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(BatchNormalization())

#conv8
model.add(Conv2DTranspose(256, (4,4), strides=2, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(256, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(256, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(256, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(BatchNormalization())
          
#conv9
model.add(Conv2DTranspose(128, (4,4), strides=2, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(128, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(128, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(BatchNormalization())

#conv10
model.add(Conv2DTranspose(128, (4,4), strides=2, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(128, (3,3), strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(Conv2D(128, (3,3), dilation_rate=1, strides=1, padding='Valid', use_bias=True))
model.add(Activation("relu"))
model.add(BatchNormalization()) 
          
model.add(UpSampling2D(size = (2,2)))         

model.add(Conv2D(2, (1,1)))
model.add(UpSampling2D(size=(2,2)))


model.compile(loss='mean_squared_logarithmic_error',
             optimizer='adam',
             metrics=['accuracy'])

```
